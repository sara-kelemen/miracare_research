```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/full_dataset_one_row_per_cycle_fw_indexing_48days.csv')
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.head()
```

```python
# keep hub_id, cycle_length, cycle_index, cycle_status, max_LH_day, max_LH_day_bw
# TRUE == data present
# FALSE == data not present
# NaN == end of cycle (as cycle length permits)
```

```python
def bool_up_to_cycle_length(df):
    hormones = ['LH', 'PDG', 'E3G']
    for hormone in hormones:
      # hormones max day 47, so range 48
        for day in range(48):
            col_name = f"day{day}_{hormone}"
            if col_name in df.columns:
                # Set NaN if day exceeds cycle length
                df.loc[day > df['cycle_length'], col_name] = np.nan
                # Set True if data is present and within cycle length
                df.loc[df[col_name].notna() & (day <= df['cycle_length']), col_name] = True
                # Set False if data is missing and within cycle length
                df.loc[df[col_name].isna() & (day <= df['cycle_length']), col_name] = False
    return df
```

```python
df = bool_up_to_cycle_length(df)
```
